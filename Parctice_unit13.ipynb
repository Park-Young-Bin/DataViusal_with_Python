## 13. 숫자 데이터를 쉽게 다루게 돕는 numpy 라이브러리

### numpy 예제

```python
import numpy as np
import matplotlib.pyplot as plt

# evenly sampled time at 200ms intervals
t = np.arange(0., 5., 0.2)    # 0부터 5 미만인 실수를 2 간격으로 추출

print(t)
print(t ** 2)
print(t ** 3)

# red dashes, blue squares and green triangles
plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^')
plt.show()
```

### 이전에 배운 내용(random 라이브러리)

```python
import matplotlib.pyplot as plt
t=[]
p2=[]
p3=[]
for i in range(0, 50, 2):    # 0부터 49까지 2 간격의 숫자 추출
    t.append(i / 10)
    p2.append((i / 10) ** 2)    # 제곱
    p3.append((i / 10) ** 3)    # 세 제곱
    
print(t)
print(p2)
print(p3)

plt.plot(t, t, 'r--', t, p2, 'bs', t, p3, 'g^')
plt.show()
```

np.random.rand(n): 0~1 사이에 있는 n개의 실수를 만든다.    
np.random.randint(a, b, n): a이상 b 미만인 정수 n개를 만든다.

ndarray에서 n은 'N차원'이라는 의미이고 array는 '배열'이라는 의미이다.  
random 라이브러리의 randint() 함수의 실수 버전(real number)이라고 생각할 수 있다.

```python
# 랜덤 추출2(random 라이브러리 사용)
# choie(a, b): 0 ~ a 미만까지의 숫자 b개를 랜덤하게 추출한다.

import numpy as np
print(np.random.choice(6, 10))   # 0부터 6까지 숫자 10개를 랜덤 추출

# 중복 제거
print(np.random.choice(10, 6, replace = False))    # 0부터 10까지 숫자 6개를 중복 없이 랜덤 추출
```

```python
# 랜덤 추출3(random 라이브러리 사용)
# choice()에 p 속성 사용

import numpy as np
print(np.random.choice(6, 10, p = [0.1, 0.2, 0.3, 0.2, 0.1, 0.1]))
```

### numpy 라이브러리를 활용해 그리프 그리기

```python
# numpy 사용한 모드
import numpy as np
import matplotlib.pyplot as plt
dice = np.random.choice(6, 10)    # 0~5까지 숫자 10개 랜덤 추출

plt.hist(dice, bins = 6)   # 가로축 구간 개수 6개
plt.show()

# Unit6에서 사용한 모드
import matplotlib.pyplot as plt
import random
dice = []
for i in range(10):
    dice.append(random.randint(0, 6))    # 0~5까지 숫자 랜덤 추출
plt.hist(dice, bins = 6)
plt.show()
```

설정한 확률에 따라 각각의 값들이 나온 것을 확인할 수 있다.

```python
# numpy를 사용한 버블 차트 그리기

import matplotlib.pyplot as plt
import numpy as np

x = np.random.randint(10, 100, 200)  # 10부터 99까지 숫자 200개를 랜덤하게 추출한다.
y = np.random.randint(10, 100, 200)
size = np.random.rand(100) * 100  # 0~1 사이에 있는 실수 100개를 랜덤하게 추출한다.

plt.scatter(x, y, s = size, c = y, cmap = 'jet', alpha = 0.7)
plt.colorbar()
plt.show()

# 오류발생: s must be a scalar, or the same size as x and y
```

```python
# Unit10에서 사용한 버블 차트 그리기

import matplotlib.pyplot as plt
import random

x = []
y = []
size = []
for i in range(200):
    x.append(random.randint(10, 100))
    y.append(random.randint(10, 100))
    size.append(random.randint(10, 100))
    
plt.scatter(x, y ,s=size, c=x, alpha = 0.7, cmap = 'jet')
plt.colorbar()
plt.show()
```

```python
import numpy as np
a = np.arange(1, 2, 0.1)    # 1 이상 2 미만 구간에서 0.1 간격으로 실수 생성
b = np.linspace(1, 2, 11)   # 1부터 2까지 11개 구간으로 나눈 실수 생성
print(a)
print(b)
```

arange(): **간격** 지정 >> 특정 간격에 해당하는 값을 생성하고 싶을 때 사용

linspace(): **구간의 개수** 지정 >> 특정 개수의 구간(n = 자연수)으로 나눈 값을 생성하고 싶을 때 사용

```python
import numpy as np
a = np.arange(-np.pi, np.pi, np.pi/10)
b = np.linspace(-np.pi, np.pi, 20)
print(a)
print(b)
```

### numpy 라이브러리를 사용하여 재미있는 버블 차트 그리기

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.randint(-100, 100, 1000)   # -100 이상 100 미만 1000개의 랜덤 값 추출
y = np.random.randint(-100, 100, 1000)   # -100 이상 100 미만 1000개의 랜덤 값 추출
size = np.random.rand(100) * 100

mask1 = abs(x) > 50   # x에 저장된 값 중 절댓값이 50보다 큰 값 걸러냄
mask2 = abs(y) > 50   # y에 저장된 값 중 절댓값이 50보다 큰 값 걸러냄
x = x[mask1+mask2]    # mask1과 mask2 중 하나라도 만족하는 값 저장
y = y[mask1+mask2]    # mask1과 mask2 중 하나라도 만족하는 값 저장
plt.scatter(x, y, s=size, c=x, cmap='jet', alpha = 0.7)
plt.colorbar()
plt.show()

# 오류발생: s must be a scalar, or the same size as x and y
```
